```python
drop_data.columns = ['Product','TOrK','T_time','S_time','Payment','Reward','A_Name','A_ID','A_gender','A_BD','B_Name','B_ID','B_gender','B_BD','A_B_Relation_Ship','S_ID','S_Name']
```

```python
ReDrop_data = drop_data.drop(['A_Name','B_Name','S_ID','S_Name'],axis = 1)
```

```python
ReDrop_data[drop_data['A_ID']!=drop_data['B_ID']].head()
```

```python
#去掉身份证不完整的A人
ReDrop_data= ReDrop_data[ReDrop_data["A_ID"].map(len)==18]
#去掉身份证不完整的B人
test = ReDrop_data.B_ID.str.isnumeric()
test = list(test.dropna().index)
ReDrop_data = ReDrop_data.loc[test]
ReDrop_data = ReDrop_data[ReDrop_data["B_ID"].map(len)==18]
#生成A的年龄信息
ReDrop_data['A_age'] =  2019 - ReDrop_data.A_BD.dt.year
#生成B的年龄信息
ReDrop_data['B_age'] =  2019 - ReDrop_data.B_BD.dt.year
#去掉年龄是119的人
ReDrop_data = ReDrop_data[ReDrop_data['A_age'] < 119 ]
```

```python
#年龄差
ReDrop_data['gap'] = ReDrop_data['A_age']-ReDrop_data['B_age']
```

```python
#是否是给自己买的
t1 = list(ReDrop_data[ReDrop_data['A_ID']!=ReDrop_data['B_ID']].index)
t2 = list(ReDrop_data[ReDrop_data['A_ID']==ReDrop_data['B_ID']].index)
ReDrop_data.loc[t1,"forSelf"] = 0
ReDrop_data.loc[t2,"forSelf"] = 1
```

```python
#是否是陕西
temp = ReDrop_data["A_ID"]
shanxi = []
notShanxi = []
for i in temp.index:
    if temp[i][0:2] != "61":
        notShanxi.append(i)
    else:
        shanxi.append(i)
ReDrop_data.loc[notShanxi,"shanxi"] = 0 # 0 表示不在陕西的人
ReDrop_data.loc[shanxi,"shanxi"] = 1 # 1 表示陕西人
```

```python
#增加attribute 年龄gap 0 无年龄gap 
#1 表示投保人比受益人大 
#2 表示投保人比受益人小

ReDrop_data.loc[ReDrop_data[ReDrop_data["forSelf"] == 1].index,"gap_Level"] = 0
```

```python
temp2 = ReDrop_data[ReDrop_data["forSelf"] != 1]["gap"] #所有投保人就是自己的人
overGap = []
lowGap = []
for i in temp2.index:
    if temp2[i] > 0:
        overGap.append(i)
    else:
        lowGap.append(i)
ReDrop_data.loc[overGap,"gap_Level"] = 1
ReDrop_data.loc[lowGap,"gap_Level"]= 2
```
